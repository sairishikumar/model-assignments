```python
import pandas as pd
#we imported pre defined librier name pandas which help to handle database or dataset contain files
```

```python
dataset=pd.read_csv("Salary_Data.csv")
#here we read the dataset file by creating a variable "dataset"
```

```python
dataset
#we can print or check the state of dataset file in stored in dataset variable
```

```python
independent=dataset[["YearsExperience"]]
independent
#here we split the table data into x-axis (input) as independent and we created as varibale as independent
```

```python
dependent=dataset[["Salary"]]
dependent
#here we split the table data into y-axis (output) as dependent and we created as varibale as dependent
```

```python
from sklearn.model_selection import train_test_split
#here we used sklearn library which can helps to perform the train test split from dataset
#then i imported train_test_split opertation from Sklearn.model_selection library
x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
#train_test_split(independent, dependent, test_size=0.30, random_state=0)here im the independent(input)& dependent (output) 
#taking the 30% data from data set for test 70% data for training purpose
#x_train,x_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0) here we split the train_test_split as x_train,x_test,y_test,y_test and we store as variables
```

```python
from sklearn.linear_model import LinearRegression
#most important part here im going to apply algorithms on my dataset so that i import linear regression alg from sklearn library
Regressor=LinearRegression()
#i created the variable as Regressor in that store the algorithms operation
Regressor.fit(x_train,y_train)
#once the regressor is ready with algorithms i implement fit method to train  X_train & y_train split data into the regressor
```

```python
weight=Regressor.coef_
weight
#i store the Refressor.coefficient method or variable into the weight
```

```python
bias=Regressor.intercept_
bias
#i store the Refressor.intercept method or variable into the bias
```

```python
y_pred=Regressor.predict(x_test)
#regressor.predict method implemnted with X_test split dataset to progress the Evluation metrics then i store into y_pred as a varible
```

```python
from sklearn.metrics import r2_score
#here i import R2_score from sklearn.metrics
r2_score=r2_score(y_test,y_pred)
#then i create r2_score method to store Y_test data & y_pred which have X_test predict information
```

```python
r2_score
#once r2_score is prgesss to run the output shows 0.9740  nearly to 1 so built model have better predit model
```

```python
import pickle
#here i import pickle library which helps to save the model to avoid retraing everytime
finalname="final assignment_linear_model.sav"
#here i created varible finalname in that "final assignment_linear_model.sav" new file name
```

```python
pickle.dump(Regressor,open(finalname,'wb'))
#here pickel.dump method in that implement regressor to open my new finalname to wb (write binary)
```

```python
loaded_model=pickle.load(open("final assignment_linear_model.sav",'rb'))
# once dump for write we need to perform pickel.load method to open my final file namne to read bias 
#all these operation store into Loaded_model
result=loaded_model.predict([[15]])
#once loaded_model.predict with given input stored into the variable as Result
```
